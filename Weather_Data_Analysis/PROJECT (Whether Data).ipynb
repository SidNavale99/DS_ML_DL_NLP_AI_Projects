```python
plt.figure(figsize=(7,7))
plt.scatter(x=data["Rel Hum_%"],y=data["Wind Speed_km/h"])
plt.ylabel("Wind Speed_km/he")
plt.xlabel("Rel Hum_%")
plt.show()
```

```python
# Histogram Plot-
```

```python
plt.figure(figsize=(7,7))
plt.hist(data["Wind Speed_km/h"],bins=10)
plt.title("Weather Condition Distribution")
plt.ylabel("Wind Speed_km/h")
plt.xlabel("Visibility_km")
plt.show
```

```python
plt.figure(figsize=(7,7))
plt.hist(data["Wind Speed_km/h"],bins=10)
plt.title("Weather Dataset distribution")
plt.ylabel("Temp_C")
plt.xlabel("Wind Speed_km/h")
plt.axvline(data["Wind Speed_km/h"].mean(),color="red")
plt.axvline(data["Wind Speed_km/h"].median(),color="yellow")
plt.legend(labels=["mean","median"])
plt.show()
```
